How much feature (Weight) we can have have by keeping low MPG

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

Loading the dataset

```python
url = 'https://raw.githubusercontent.com/tusarrpatel/Collab/main/auto-mpg.csv'
auto = pd.read_csv(url)
auto.head
```

Initiating two variables. X to represent the feature and y to represent the target

```python
X = auto[['weight']]
y = auto['mpg']
```

Create the LinearRegression class and train the model

```python
MPG_Reg = LinearRegression()
MPG_Reg.fit(X,y)
```

Again the plotter

```python
plt.figure(figsize=(10,10))
plt.scatter(auto['weight'],auto['mpg'])
plt.plot(X,MPG_Reg.predict(X),c='red')
plt.title('Miles per gallon VS weight of car')
plt.xlabel('Weight of car')
plt.ylabel('Miles per gallon')
plt.show()
```
